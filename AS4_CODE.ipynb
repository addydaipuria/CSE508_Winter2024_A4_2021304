```python
import pandas as pd
import re

def cleaning(s):
    s = str(s)
    s = re.sub(r'\s\W', ' ', s)
    s = re.sub(r'\W,\s', ' ', s)
    s = re.sub(r'\d+', '', s)
    s = re.sub(r'\s+', ' ', s)
    s = re.sub(r'[!@#$_]', '', s)
    s = s.replace("co", "")
    s = s.replace("https", "")
    s = s.replace(r"\[\w\*", " ")
    return s


df = pd.read_csv('C:/Users/adida/OneDrive/Desktop/IR/amazonfood/Reviews.csv')

df = df.dropna()
df['Text'] = df['Text'].apply(cleaning)
df['Summary'] = df['Summary'].apply(cleaning)

df['Combined'] = df['Text'] + ' TL;DR ' + df['Summary']

df.to_csv('C:/Users/adida/OneDrive/Desktop/IR/amazonfood/clean_Reviews.csv', index=False)
```

```python
from torch.utils.data import Dataset
import torch

class ReviewDataset(Dataset):
    def __init__(self, reviews, tokenizer, max_length=100):
        self.reviews = reviews
        self.tokenizer = tokenizer
        self.max_length = max_length
        self.tokenizer.pad_token_id = self.tokenizer.eos_token_id

    def __len__(self):
        return len(self.reviews)

    def __getitem__(self, idx):
        review = self.reviews[idx]

        tokens = self.tokenizer.encode(review + self.tokenizer.eos_token )

        if len(tokens) > self.max_length:
            tokens = tokens[:self.max_length]
        else:
            padding_length = self.max_length - len(tokens)
            tokens = tokens + [self.tokenizer.pad_token_id] * padding_length

        labels = tokens[1:] + [self.tokenizer.pad_token_id]

        return {
            'input_ids': torch.tensor(tokens),
            'labels': torch.tensor(labels)
        }
```

```python
from rouge_score import rouge_scorer

scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
test_review = test_df['Text'].tolist()  
test_summary = test_df['Summary'].tolist()  

generated_summaries = []

for review in test_review:
    inputs = tokenizer(review, return_tensors='pt', max_length=100, truncation=True)
    input_ids = inputs.input_ids.to(device)
    with torch.no_grad():
        output = model.generate(input_ids, max_new_tokens=100, num_return_sequences=1, do_sample=True)
    generated_summary = tokenizer.decode(output[0], skip_special_tokens=True)
    generated_summaries.append(generated_summary)

rouge_scores = []
for generated_summary, actual_summary in zip(generated_summaries, test_summary):
    scores = scorer.score(generated_summary, actual_summary)
    rouge_scores.append(scores)

print("ROUGE Scores:")
for i, scores in enumerate(rouge_scores):
    print(f"Review {i+1}:")
    print(scores)
```
